```python
import sqlite3
from sqlite3 import Error
 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn


```

```python
''' Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members!'''
  
    cur = conn.cursor()
   
         
    query1 =  """SELECT f.name,  sum(f.guestcost*b.slots + f.membercost*b.slots) as total_revenue
                FROM Facilities as f
                LEFT JOIN Bookings as b
                WHERE (b.starttime BETWEEN '2012-09-14 00:00:00' AND '2012-09-14 23:59:00' 
                AND f.guestcost * b.slots + f.membercost * b.slots < 1000)
                GROUP BY f.name
                ORDER BY total_revenue Desc
                ;"""
              
    cur.execute(query1)
 
    rows = cur.fetchall()

    for row in rows:
            print(row)
    
   
#..................OUTPUT......................
('Massage Room 2', 12406.199999999988)
('Massage Room 1', 12406.199999999988)
('Tennis Court 2', 4140)
('Tennis Court 1', 4140)
('Squash Court', 2898.0)
('Badminton Court', 2139.0)
('Table Tennis', 690)
('Snooker Table', 690)
('Pool Table', 690)
```
